# Exercise 2

### Create box plots for all of the numercial variables in Fisher's Iris Data Set

```python
# Resource: https://matplotlib.org/stable/plot_types/index
# Resource: http://seaborn.pydata.org/introduction.html#visualizing-dataset-structure
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
   
data = pd.read_csv('Iris-Data-Set.csv')

# Show the boxplot of Irish Data Set
# Show boxplot for all 4 columns
data.boxplot()

plt.title('Boxplot of each variable in the Iris Data Set')
plt.show()
```

```python
# Split the boxplot for each column in the Iris Data Set

# Apply black background
plt.style.use("dark_background")

# Apply different colours per each class
colors = {"yellow", "orange", "red"}

# Show boxplot for Petal Width
sns.boxplot(x=data['class'], y=data['petal_width'], palette=colors)

plt.title('Petal Width for 3 Iris Classes')
plt.show()
```

```python
# Show boxplot for Petal Lenght
sns.boxplot(x=data['class'], y=data['petal_length'], palette=colors)

plt.title('Petal Length for 3 Iris Classes')
plt.show()
```

```python
# Show boxplot for Sepal Width
sns.boxplot(x=data['class'], y=data['sepal_width'], palette=colors)

plt.title('Sepal Width for 3 Iris Classes')
plt.show()
```

```python
# Show boxplot for Sepal Lenght
sns.boxplot(x=data['class'], y=data['sepal_length'], palette=colors)

plt.title('Sepal Length for 3 Iris Classes')
plt.show()
```
